```python
import random
import matplotlib.pyplot as plt

#intervalo dos números aleatórios e o tamanho da lista
start = 1
end = 100
list_size = 1_000_000

# map com random.randint para gerar uma lista de números aleatórios
random_numbers = list(map(lambda _: random.randint(start, end), range(list_size)))

# histograma dos números gerados
plt.hist(random_numbers, bins=range(start, end+2), edgecolor='black')
plt.title('Histograma de números aleatórios')
plt.xlabel('Números aleatórios')
plt.ylabel('Quantidade')
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def generate_random_lists(list_size, start, end):
    """
    Gera duas listas de números aleatórios com o mesmo tamanho.
    """
    list1 = [random.randint(start, end) for _ in range(list_size)]
    list2 = [random.randint(start, end) for _ in range(list_size)]
    return list1, list2

def plot_sum_of_random_numbers(list1, list2):
    """
    Plota o histograma da soma de elementos correspondentes de duas listas.
    """
    sums = [x + y for x, y in zip(list1, list2)]
    plt.hist(sums, bins=range(min(sums), max(sums) + 2), edgecolor='black')
    plt.title('Histograma da Soma 2 de Números Aleatórios')
    plt.xlabel('Soma de 2 números')
    plt.ylabel('Quantidade')
    plt.grid(True)
    plt.show()

# intervalo dos números aleatórios e o tamanho das listas
start = 1
end = 50
list_size = 1_000_000

# duas listas de números aleatórios
list1, list2 = generate_random_lists(list_size, start, end)

# Plote o grafico
plot_sum_of_random_numbers(list1, list2)
```

```python
import random
import matplotlib.pyplot as plt

def generate_random_lists(list_size, start, end):
    """
    Gera três listas de números aleatórios com o mesmo tamanho.
    """
    list1 = [random.randint(start, end) for _ in range(list_size)]
    list2 = [random.randint(start, end) for _ in range(list_size)]
    list3 = [random.randint(start, end) for _ in range(list_size)]
    return list1, list2, list3

def plot_sum_of_random_numbers(list1, list2, list3):
    """
    Plota o histograma da soma de elementos correspondentes das três listas.
    """
    sums = [x + y + z for x, y, z in zip(list1, list2, list3)]
    plt.hist(sums, bins=range(min(sums), max(sums) + 2), edgecolor='black')
    plt.title('Histograma da Soma de 3 Números Aleatórios')
    plt.xlabel('Soma de 3 números')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()

# intervalo dos números aleatórios e o tamanho das listas
start = 1
end = 33
list_size = 1_000_000

# três listas de números aleatórios
list1, list2, list3 = generate_random_lists(list_size, start, end)

# Plote do grafico
plot_sum_of_random_numbers(list1, list2, list3)
```

```python
import random
import matplotlib.pyplot as plt

def generate_random_lists(list_size, start, end):
    """
    Gera cinco listas de números aleatórios com o mesmo tamanho.
    """
    list1 = [random.randint(start, end) for _ in range(list_size)]
    list2 = [random.randint(start, end) for _ in range(list_size)]
    list3 = [random.randint(start, end) for _ in range(list_size)]
    list4 = [random.randint(start, end) for _ in range(list_size)]
    list5 = [random.randint(start, end) for _ in range(list_size)]
    return list1, list2, list3, list4, list5

def plot_sum_of_random_numbers(list1, list2, list3, list4, list5):
    """
    Plota o histograma da soma de elementos correspondentes das cinco listas.
    """
    sums = [x + y + z + w + v for x, y, z, w, v in zip(list1, list2, list3, list4, list5)]
    plt.hist(sums, bins=range(min(sums), max(sums) + 2), edgecolor='black')
    plt.title('Histograma da Soma de 5 Números Aleatórios')
    plt.xlabel('Soma de 5 números')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()

# Defina o intervalo dos números aleatórios e o tamanho das listas
start = 1
end = 20
list_size = 1_000_000

# Gere as cinco listas de números aleatórios
list1, list2, list3, list4, list5 = generate_random_lists(list_size, start, end)

# Plote a soma dos números aleatórios das cinco listas
plot_sum_of_random_numbers(list1, list2, list3, list4, list5)
```

```python
import random
import matplotlib.pyplot as plt

def generate_random_lists(list_size, start, end):
    """
    Gera dez listas de números aleatórios com o mesmo tamanho.
    """
    random_lists = []
    for _ in range(10):
        random_list = [random.randint(start, end) for _ in range(list_size)]
        random_lists.append(random_list)
    return random_lists

def plot_sum_of_random_numbers(random_lists):
    """
    Plota o histograma da soma de elementos correspondentes das dez listas.
    """
    sums = [sum(x) for x in zip(*random_lists)]
    plt.hist(sums, bins=range(min(sums), max(sums) + 2), edgecolor='black')
    plt.title('Histograma da Soma de 10 Números Aleatórios')
    plt.xlabel('Soma de 10 números')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()

# Defina o intervalo dos números aleatórios e o tamanho das listas
start = 1
end = 100
list_size = 1_000_000

# Gere as dez listas de números aleatórios
random_lists = generate_random_lists(list_size, start, end)

# Plote a soma dos números aleatórios das dez listas
plot_sum_of_random_numbers(random_lists)
```
